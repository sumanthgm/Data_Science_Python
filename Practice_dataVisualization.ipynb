```python
# Add marker
sns.lineplot(data=frmi,x="timepoint",y="signal",hue="region",style="region",marker=True)
```

**Joint Ploat**

---



```python
# Joint ploat mainly used to create join mutiple plots
sns.jointplot(data=data,x="engine_size",y="horsepower")
```

```python
# Colour scalling using in joint plot operation use (kind hex) function
sns.jointplot(data=data,x="engine_size",y="horsepower",kind="hex")
```

```python
# Colour scalling using in joint plot operation use (kind kde) function
sns.jointplot(data=data,x="engine_size",y="horsepower",kind="kde")
```

```python
# kind-reg used to find the best curve fits plot
sns.jointplot(data=data,x="engine_size",y="horsepower",kind="reg")
```

```python
# kind-reg used to find the best curve fits plot deffernt data points
sns.jointplot(data=data,x="price",y="city_mpg",kind="reg")
```

**Strip Ploat**

---



```python
# Strip ploat is same as scatter plot datapoints strip with visualization
plt.figure(figsize=(15,7))
sns.stripplot(data=data,x="body_style",y="engine_size",hue="body_style")
```
