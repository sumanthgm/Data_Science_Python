**Line Graph**

```python
# Linegraph Image
Image("/content/drive/MyDrive/Data - Science (PES)/Sem 1/Python /Olympus data/data_Visualization/Line_plot.png")
```

```python
flights=sns.load_dataset("flights")
sns.lineplot(data=flights,x="month",y="passengers")
```

```python
# We can swith off confidence interval by suing ci parameter to false
sns.lineplot(data=flights,x="month",y="passengers",ci= False)
# sns.lineplot(data=flights,x="month",y="passengers",ci= False)
```

```python
# We can had deffernce parameters in dataset ading hue colour parrameter
sns.lineplot(data=flights,x="month",y="passengers",ci= False, hue="year")
```

```python
# Change the style in deffrent parametrs by adding style function
frmi=sns.load_dataset("fmri")
#  creting line plt.plot
sns.lineplot(data=frmi,x="timepoint",y="signal",hue="region",style="region",ci=False)



```

```python
# Add marker
sns.lineplot(data=frmi,x="timepoint",y="signal",hue="region",style="region",marker=True)
```

```python
# Chnage legend or location in the dataset
plt.figure(figsize=(15,7))
sns.lineplot(data=flights,x="month",y="passengers",hue="year")
plt.ylabel("Number of passengers")
plt.legend(bbox_to_anchor=[1,1])
```
